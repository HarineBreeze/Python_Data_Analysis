## 一、干净数据的标准
#### ① 一列只反映单个数据（错误示范：gender_birthday） ② 数值大小符合实际（错误示范：身高3.1）
#### ④ 表示唯一信息的数据不重复出现  ④ 关键信息不能缺失 （错误示范：有其他数据但姓名为NaN空值）

```python
import pandas as pd
studens_grade = pd.read_csv("D:/PythonProject/PyDataBase/studens_grade.csv",index_col="id")
studens_grade
```

## 二、列表结构评估

#### .rename(index={"原名":"新名"}) 重命名指定索引
#### .rename(columns={"原名":"新名"})重命名指定列名

```python
# 此处一定要对DataFrame重新赋值，不然只会呈现改变效果，不实际改变原表（也可以在{"原名":"新名"},inplace=True的方式达到同样效果）
studens_grade = studens_grade.rename(columns={"身高height":"身高"}).rename(index={"三零一":"301"}) 
studens_grade
```

#### .rename(columns=str.upper) 让所有列名变为大写，此处str.upper是可更换的函数
#### .reset_index() 将自定义的索引还原为初始0,1,2...，同时自定义的索引成为单独的一列数据
#### .sort_index(axis=0) 对索引以列为单位进行升序排序

```python
studens_grade.sort_index(inplace=True)
studens_grade
```

## 三、丢失数据
#### 1.人工填入缺失值 2.不处理缺失值 3.删除缺失值的行 4.用例如平均数、众数等值填充代替缺失值

#### .info() 展示列表中所有数据的一些统计信息 （用于清洗数据时评估丢失数据）

```python
studens_grade.info()
```

#### .isnull().sum() 计算出列表中各列名下空值的总数（True被视为1，False视为0）

```python
studens_grade.isnull().sum()
```

#### 方法同上，但是使用["列名"]来提取特定列进行筛查

```python
studens_grade["姓名"].isnull().sum()
```

#### 变量名 [ 变量名 [列名] . isnull () ] 通过如上方法找到丢失数据后，提取丢失数据的行
#### 若要筛选多个条件，则使用括号对每个条件进行分割，例如：变量名[(变量名["列1"]>0) & (变量名["列1"]!=5) & (变量名["列3"]==0)]

```python
studens_grade[studens_grade["姓名"].isnull()]
```

#### 为缺失值103,姓名补充数据

```python
studens_grade.loc["103","姓名"]="小菲"
studens_grade
```

## 四、重复数据
#### 1.删除重复数据
#### .duplicated ( subset = ["列名1","列名2"] ) 提取列表中列名1和列名2同时重复的数据

```python
studens_grade.duplicated(subset=["姓名","班级"])
```

#### 经过上面的筛查发现301返回True(重复值)，使用变量名[变量名.duplicated(subset=["变量1","变量2"])]，将相应的重复值提取出来查看

```python
studens_grade[studens_grade.duplicated(subset=["姓名","班级"])]
```

#### .drop_duplicates(subset="列名",keep="first") 删除重复行，只保留第一个（first可换为last,意为保留最后出现的重复值）

```python
studens_grade.drop_duplicates(subset='姓名', keep='first',inplace=True)
studens_grade
```

## 五、无效/错误数据
#### 1.删除无效数据 2.人工修正无效数据 3.用平均数或众数等进行填充

计数（count）、平均值（mean）、标准差（std）、最小值（min）、百分位数（如 25%、50%、75%）、最大值（max）

## 六、改错实践

```python
data_cleaning = pd.read_csv("D:/PythonProject/PyDataBase/mistake_data_cleaning.csv")
data_cleaning
```

### 1.拆分一列存在多数据
#### 变量名[["新列1","新列2"]] = 变量名["拆分列名"].split("拆分位置",expand=True) 

```python
data_cleaning[["人口","面积"]] = data_cleaning["人口密度"].str.split("/",expand=True)
# 删除"人口密度"列，且通过inplace=True，直接改变原列表结果
data_cleaning.drop("人口密度",axis=1,inplace=True)
data_cleaning
```

### 2.合并两列数据
#### 变量名["列1"].str.cat(data_cleaning["列2"])

```python
data_cleaning["管理者姓"].str.cat(data_cleaning["管理者名"])
#data_cleaning = data_cleaning.rename(columns={"管理者姓":"管理者姓名"})
data_cleaning
```

### 3.将宽数据转为长数据
#### pd.melt(变量名,id_vars=["不希望转换的列1"，"不希望转换的列N"],var_name="新形成的列命名",value_name="数据的列名")

### 4.将指指定列名下的一行单元格的数据拆分成多行
#### 变量名.explode("列名") 

### 5.删除缺失值所在行
#### 变量名.dropna(subset=["列名"]) 括号内内容可有可无，如有则意味只对指定列执行删除缺失值行操作

```python
data_cleaning.dropna(subset=["平均年龄"],inplace=True)
data_cleaning
```

### 6.填充缺失值
#### 变量名["列名"] = 变量名.fillna("列名".mean())

```python
mean_value = data_cleaning["收入评级"].mean()
data_cleaning["收入评级"] = data_cleaning["收入评级"].fillna(mean_value)
data_cleaning
```

### 7.将指定值替换
#### 变量名[列名].replace("要替换的值","替换后的值") 此处要替换的值可以为一个列表["1","2","3",...],也可以整体为一个字典（键替换为值）

```python
replace_dict = {"A级":"优等","B级":"中等"}
data_cleaning["科技评级"]= data_cleaning["科技评级"].replace(replace_dict)
data_cleaning
```

### 8.转换特定列的的数据类型
#### 变量名["列名"].astype(数据类型)

```python
data_cleaning["平均年龄"] = data_cleaning["平均年龄"].astype(int)
data_cleaning["科技评级"] = data_cleaning["科技评级"].astype("category")
type("平均年龄")
data_cleaning
```

### 9.将不符合条件的值清除
#### 原理就是筛选出符合条件的值，然后再对该变量重新赋值为该筛选结果

```python
data_cleaning = data_cleaning[data_cleaning["管理者名"].str.len()==2]
# 将列表只保留"管理者名"符合字符串长度为2的数据
data_cleaning
```
